```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('insurance.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['sex'].value_counts()
```

```python
dict_sex = {'male':1,'female':2}
dict_sex

# 1 for male
# 2 for female
```

```python
df['sex'] = df['sex'].map(dict_sex)
df.head()
```

```python
df['smoker'].value_counts()
```

```python
dict_smoker = {'yes':1,'no':0}
dict_smoker
```

```python
df['smoker'] = df['smoker'].map(dict_smoker)
df.head()
```

```python
df['region'].value_counts()
```

```python
dict_region = {'northeast':1,'northwest':2,'southeast':3,'southwest':4}
dict_region
```

```python
df['region'] = df['region'].map(dict_region)
df.head()
```

```python
# all the columns are now numerical data columns
```

```python
df.head(2)
```

```python
# determining x variables and y variable

x = df.drop('charges',axis=1)
y = df[['charges']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
# scikit-learn
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
# 80% training size
```

```python
from sklearn.linear_model import LinearRegression
# algorithm import
```

```python
lnr = LinearRegression()
```

```python
lnr.fit(x_train,y_train)   # model training
```

```python
x_train.shape
```

```python
x_train.isnull().sum()
```

```python
x_train
```

```python
lnr.score(x_train,y_train)*100
# training score
```

```python
lnr.score(x_test,y_test)*100
# testing score
```

```python
# prediction
prediction = lnr.predict(x_test)
```

```python
y_test
# original charges (not predicted charges)
```

```python
y_test['prediction'] = prediction 
```
